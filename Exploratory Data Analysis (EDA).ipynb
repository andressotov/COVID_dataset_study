```python
# Total number of samples
nrows = df.shape[0]
```

```python
died = {True: 'died', False: 'alive'}

def contar(label, atrib):
    ld = []
    for val in atrib.keys():
        # print('\natrib:', val)
        for d in died.keys():
            dic = {label: atrib[val]}
            # print('died or alive d=', d, 'died[d]', died[d])
            dfr = df.loc[(df['DATE_DIED'] == d) & (df[label] == val)]
            dic['died/alive'] = died[d]
            dic['amount'] = dfr.shape[0]
            dic['percentage'] = (dfr.shape[0] / nrows) * 100
            # print('Amount', died[d], ':', dfr.shape[0], label, atrib[val], '%:', (dfr.shape[0]/nrows)*100)
            # pprint(dic)
            ld.append(dic)
    return ld
```

```python
from pprint import pprint

sex = {1: 'female', 2: 'male'}
res = contar('SEX', sex)
# pprint(res)

df_sex = pd.DataFrame(res)
print(df_sex)
```

```python
llist = df_sex.values.tolist()
newl = [ [l1+'_'+l2, n1, n2]  for [l1, l2, n1, n2] in llist]
[x, y1, y2] = zip(*newl)
x = np.asarray(x)
print(x)
y1 = np.asarray(y1)
print(y1)
y2 = np.asarray(y2)
print(y2)
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, (ax1, ax2) = plt.subplots(2, 1, constrained_layout=True, sharex=True)
ax1.set_title('Amount')
ax1.bar(x, y1, color='blue')
ax2.set_title('Percentage')
ax2.bar(x, y2, color='green')
plt.show()
```
